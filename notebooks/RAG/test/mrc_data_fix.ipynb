```python
# 기존 klue-mrc 데이터가 context와 매칭되는 ducument의 id를 가지고 있지 않아서, document id를 추가하는 코드입니다.
```

```python
import os

import pandas as pd
```

```python
context_data = pd.read_csv("../data/question,context.csv", encoding="utf-8")
document_data = pd.read_csv("../data/docs.csv", encoding="utf-8")
```

```python

if 'id' not in document_data.columns:
    document_data['id'] = range(len(document_data))


context_data['document_id'] = None

for i, row in context_data.iterrows():
    context = row["context"]
    question = row["question"]
    
 
    match = document_data[document_data['document'] == context]
    if not match.empty:
        context_data.at[i, 'document_id'] = match.iloc[0]['id']
        context_data.at[i, 'cid'] = i
        print("context and question are same")
        print("context: ", context)
        print("question: ", question)
        print("index: ", i)
        print("context_data_id: ", i)
        print("document_id: ", match.iloc[0]['id'])
        print("\n")
```

```python
null_data = context_data[context_data['document_id'].isnull()]

print("null data: ", null_data)
```

```python
## context_data에서 document_id가 None인 것들만 추출
context_data = context_data[context_data['document_id'].notnull()]

context_data.to_csv("../data/klue-mrc-add-id.csv", index=False, encoding="utf-8")
```
